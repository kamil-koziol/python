```python
import os
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
IMG_PATH = "./rps-test-set/"
CATEGORIES = ['rock','paper','scissors']
test_data = []

for category in CATEGORIES:
    path = os.path.join(IMG_PATH, category)
    class_num = CATEGORIES.index(category)
    for img in os.listdir(path):
        try:
            img_array = cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE)
            new_array = cv2.resize(img_array, (300,300))
            test_data.append([new_array, class_num])
        except Exception as e:
            pass
```

```python
x = []
y = []

for feature, label in test_data:
    x.append(feature)
    y.append(label)

x = np.array(x).reshape(-1,300,300,1) # 1 jest jako grayscale, 3 kolor
y = np.array(y)
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(64, (3,3), input_shape=(300,300,1), activation=tf.nn.relu),
    tf.keras.layers.MaxPooling2D(2,2),
    
    tf.keras.layers.Conv2D(64, (3,3), activation=tf.nn.relu),
    tf.keras.layers.MaxPooling2D(2,2),
    
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Flatten(),
    
    tf.keras.layers.Dense(256, activation=tf.nn.relu),
    tf.keras.layers.Dense(3, activation=tf.nn.softmax)
])

model.compile(
    optimizer=tf.keras.optimizers.RMSprop(),
    loss=tf.keras.losses.sparse_categorical_crossentropy,
    metrics = ['accuracy'])
```

```python
model.load_weights('./rpsTrain/cp1.cpkt')
```

```python
model.evaluate(x,y)
```

```python
def prepare_image(path):
    img_array = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    new_array = cv2.resize(img_array, (300,300))    
    return new_array

img = prepare_image("./rpsTestImg/papier.JPG")
img = img.reshape(-1,300,300,1)
img = img/255
```
