```python
from __future__ import absolute_import, division, print_function, unicode_literals

import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

```python
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()
```

```python
# zmniejszenie ilości zdjęć
train_labels = train_labels[:1000]
test_labels = test_labels[:1000]

# teraz dajemy reshape aby potem nie używać Flatten
train_images = train_images[:1000].reshape(-1, 28 * 28) / 255.0
test_images = test_images[:1000].reshape(-1, 28 * 28) / 255.0
```

```python
def create_model():
    model= tf.keras.models.Sequential([
        keras.layers.Dense(512, activation=keras.activations.relu, input_shape=(784,)), # input layer
        keras.layers.Dense(10, activation=keras.activations.softmax) # 10 bo jest 10 cyfr
    ])
    
    model.compile(
        optimizer=keras.optimizers.Adam(),
        loss=keras.losses.sparse_categorical_crossentropy,
        metrics=["accuracy"])
    
    return model
```

```python
model = create_model()
```

```python
# zapisywanie wag modelu
checkpoint_path = "training_1/cp.ckpt"
checkpoint_dir = os.path.dirname(checkpoint_path)

cp_callback = keras.callbacks.ModelCheckpoint(checkpoint_path, save_weights_only=True, verbose=1)

model.fit(train_images, train_labels, epochs=10,
         validation_data= (test_images, test_labels),
         callbacks=[cp_callback])
```

```python
# tworzenie pustego nowego modelu - niewytrenowanego
newModel = create_model()

newModel.evaluate(test_images, test_labels)
```

```python
# wczytanie wag
newModel.load_weights(checkpoint_path)
newModel.evaluate(test_images, test_labels)
```
