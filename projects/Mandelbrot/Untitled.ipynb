```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mandelbrot(Re, Im, max_iter, z=0j):
    c = complex(Re, Im)
    
    for i in range(max_iter):
        z = z*z + c
        if z.real*z.real + z.imag*z.imag >= 4:
            return i
        
    return max_iter
```

```python
columns = 500
rows = 500
max_iterations = 100
zoom = 2

screen = np.zeros((rows, columns))
for rowIndex, Re in enumerate(np.linspace(-zoom, zoom, num=rows)):
    for columnIndex, Im in enumerate(np.linspace(-zoom, zoom, num=columns)):
        screen[rowIndex, columnIndex] = mandelbrot(Re, Im, max_iterations)

# Julia Set
# screen = np.zeros((rows, columns))
# Re = 0.285
# Im = 0.01
# for row in range(rows):
#     for column in range(columns):
#         screen[row, column] = mandelbrot(Re, Im, max_iterations)
```

```python
plt.figure(dpi=100)
plt.imshow(screen, cmap="hot")
plt.xlabel("Re")
plt.ylabel("Im")
plt.show()
```
